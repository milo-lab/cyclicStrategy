```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
from sklearn.linear_model import LinearRegression
from IPython.core.debugger import set_trace
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

def init_fig(aspect=(4,4),fontsize=12):
    plt.rc('font', family='sans-serif')
    plt.rc('xtick', labelsize=fontsize)
    plt.rc('ytick', labelsize=fontsize)
    plt.rc('axes', labelsize=fontsize)
    params = {'legend.fontsize': 'larger',
              'legend.handlelength': 2}
    plt.rcParams.update(params)

    fig = plt.figure(figsize=aspect,dpi=300)
    ax = fig.add_subplot(1, 1, 1)
    return ax,fig
```

### Model definition

```python
def SEIR_erlang(X,n_groups,n_comp,Rt,L_days,I_days,delta,dt=0.1,rho=1,W=1.5,L=0.6):
  
  # X - matrix with N groups and M compartments
  # n_comp - number of compartments - k of Erlang
  # n_groups - number of shifts
  # Rt - vector of R values for each group
  # L_days - latent period
  # I_days - infectious period
  # delta - the fraction of spillover
  # Rf - the rate of infections from one group to the other
  # dt - time step
  # W - R for work days
  # L - R for lockdown days
  
  # if we are simulating only one group rho has to be equal to one
  if n_groups == 1:
    rho=1

  dSdt = np.zeros([n_groups])
  dEdt = np.zeros([n_groups,n_comp])
  dIdt = np.zeros([n_groups,n_comp])
  dRdt = np.zeros([n_groups])
  
  # run time step for each group
  for group in np.arange(n_groups):
    # load state from input
    x = X[group,:].squeeze()
    E = x[:n_comp]
    I = x[n_comp:2*n_comp]
    S = x[-2]
    R= x[-1]
    
    # define the number of infected people in other groups and their R
    I_other = np.delete(X,group,axis=0)[:,n_comp:2*n_comp].sum(axis=1)
    R_other = np.delete(Rt,group,axis=0)
    
    
    d=0.01; # The rate at which I becomes R
    s=1/L_days; # rate at which E becomes I
    g=1/I_days; # rate at which I becomes R
    
    if Rt[group] == L:
      fun = (n_groups)*Rt[group]*g*S*I.sum() + (rho*delta*(R_other-L)*g*S*I_other[:,np.newaxis]).sum() + (rho*delta**2*(R_other-L)*g*S*I.sum()).sum()
    else:
      fun = (n_groups*L + rho*(Rt[group]-L))*g*S*I.sum() + (rho*delta*(W-R_other)*g*S*I_other[:,np.newaxis]).sum()
    #We assume S is not changing
    dSdt[group] = 0
    dEdt[group,0] = fun-n_comp*s*E[0]
    for i in range(1,n_comp):
        dEdt[group,i] = n_comp*s*(E[i-1]-E[i])
    
    dIdt[group,0] = n_comp*(s*E[-1]-g*I[0])
    for i in range(1,n_comp):
        dIdt[group,i] = n_comp*g*(I[i-1]-I[i])

    dRdt[group]=np.array([n_comp*g*I[-1]]);
  
  dydt=np.concatenate([dEdt*dt,dIdt*dt,dSdt[:,np.newaxis]*dt,dRdt[:,np.newaxis]*dt],axis=1)
  return dydt

def test_time(t,period):
  te = lambda x: [(x>=y[0]) & (x<y[1]) for y in period]
  return any(te(t))


def start_sim(n_cont=2,R0=2.4,R_L=0.6,t_on=14,t_off=28,dt=0.1,all=False,L_days=3,I_days=4):
  # Start the simulation with a t_on days of free growth with R of R0 and the t_off days of lockdown with R_L
  n_groups = 1
  rh = 1
  R = np.ones(n_groups)
  delta = 0

  x_e = np.zeros((1,n_groups,n_cont*2+2))
  x_e[:,:,:2*n_cont] = 1e-3
  x_e[:,:,-2] = 1/n_groups

  # run for t_on days with unmitigated R
  Rt=R*R0
  for t in np.arange(0,t_on,dt):
      dx = SEIR_erlang(x_e[-1,:,:],n_groups,n_cont,Rt,L_days,I_days,delta,dt,rh,R0,R_L)
      x_e = np.append(x_e,(x_e[-1,:]+dx)[np.newaxis,:],axis=0)

  # run t_off weeks of lockdown
  Rt=R*R_L
  for t in np.arange(0,t_off,dt):
      dx = SEIR_erlang(x_e[-1,:,:],n_groups,n_cont,Rt,L_days,I_days,delta,dt,rh,R0,R_L)      
      x_e = np.append(x_e,(x_e[-1,:]+dx)[np.newaxis,:],axis=0)
  if all:
    return x_e.copy()
  else:
    return x_e[-1:,:,:]

def sim_groups(R_W,R_L,rh,t_cycle,dt,periods,n_groups,delta=0.1,plot=False,all=False,full=False,x0=None,tmax=125,L_days=3,I_days=4,n_cont=2):
  # simulate n groups with a work-lockdown schedule define in the variable periods
  R = np.ones(n_groups)
  
  # initialize epidemic
  if x0 is None:
    x_e = np.zeros((1,n_groups,n_cont*2+2))
    x_start = start_sim(all=all,L_days=L_days,I_days=I_days,n_cont=n_cont)
    x_e[:,:,:] = x_start[-1,:,:]
    x_e[:,:,-2] = x_e[:,:,-2]/n_groups
  else:
    x_e = x0
  
  Rt = R*R_L
  t_sim = np.arange(0,tmax,dt)
  for t in t_sim:
    # set the state (work/lockdown) for each group
    for group in range(n_groups):
      if test_time(np.mod(t,t_cycle),periods[group]):
        Rt[group]=R_W
      else:
        Rt[group]=R_L      
    
    # run a simulation timestep
    dx = SEIR_erlang(x_e[-1,:,:],n_groups,n_cont,Rt,L_days,I_days,delta,dt,rh,R_W,R_L)
    x_e = np.append(x_e,(x_e[-1,:]+dx)[np.newaxis,:],axis=0)
  

  x_e_plot = x_e[:,:,n_cont:2*n_cont].sum(axis=(1,2))/x_e[0,:,n_cont:2*n_cont].sum()
  x_t = np.arange(0,x_e_plot.shape[0]*dt,dt)
  
  # fit a linear model to the log transformed time-series of the infected population
  linfit = LinearRegression()  
  nf = int(1/dt)
  X = (x_t[20*nf:]).reshape(-1,1)
  Y = np.log(x_e_plot[20*nf:]).reshape(-1,1)
  linfit.fit(X, Y) 
#   linfit.fit((x_t).reshape(-1,1), np.log(x_e_plot.reshape(-1,1))) 
  
  if plot:
    if any([all,full]):
      if x0 is None:
        x_all = np.zeros([x_start.shape[0]+x_e.shape[0],n_groups,x_start.shape[2]])
        for g in range(n_groups):
          x_all[:,g:g+1,:] = np.concatenate([x_start,x_e[:,g:g+1,:]],axis=0)
      else:
        x_all = x_e
      if full:
        x_all_plot = x_all
      else:
        x_all_plot = x_all[:,:,n_cont:2*n_cont].sum(axis=(2))/x_all[0,:,n_cont:2*n_cont].sum()
      x_t_all = np.arange(0,x_all_plot.shape[0]*dt,dt)
      return x_t_all,x_all_plot
    else:
      return linfit,x_t,x_e_plot
  else:
    return linfit.coef_

def sim_bang(R_W,R_L,rh,t_off,dt,n_groups,delta=0.1,t_on=14,full=False,L_days=3,I_days=4):
  # simulate a upper and lower threshold for staring and stopping lockdown with n groups
  Rt = np.ones(n_groups)
  
  # initialize epidemic
  x_bang = start_sim(all=True,t_off=t_off)
  x_bang[:,:,-2] = x_bang[:,:,-2]/n_groups

  l_time = []
  w_time = []
  
  # set upper and lower thresholds
  c_t = 2
  c_l = x_bang[-1,:,n_cont:2*n_cont].sum()/x_bang[0,:,n_cont:2*n_cont].sum()*1.4

  Rt[:]=R_W
  
  # run simulation for a total of 168 days
  t_sim_norm = np.arange(0,168-t_on-t_off,dt)
  for t in t_sim_norm:
    # if the number of infected people has crossed the threshold, start locdown
    if x_bang[-1,:,n_cont:2*n_cont].sum()/x_bang[0,:,n_cont:2*n_cont].sum() >= c_t:
      if (Rt[:]==R_W).all():
          l_time.append(t)
      Rt[:]=R_L
    # if the number of infected people has decreased below the lower threshold, release lockdown
    elif x_bang[-1,:,n_cont:2*n_cont].sum()/x_bang[0,:,n_cont:2*n_cont].sum() <= c_l:
      if (Rt[:]==R_L).all():
          w_time.append(t)
      Rt[:]=R_W  
    
    # run a simulation timestep
    dx = SEIR_erlang(x_bang[-1,:,:],n_groups,n_cont,Rt,L_days,I_days,delta,dt,rh,R_W,R_L)
    x_bang = np.append(x_bang,(x_bang[-1,:]+dx)[np.newaxis,:],axis=0)
  
  if full:
    x_e_plot = x_bang
  else:
    x_e_plot = x_bang[:,:,n_cont:2*n_cont].sum(axis=(1,2))/x_bang[0,:,n_cont:2*n_cont].sum()
  x_t = np.arange(0,x_e_plot.shape[0]*dt-dt,dt)
  return x_t,x_e_plot,l_time,w_time

def get_s_to_R(L_days,I_days,range,k=2):
  R_range = np.arange(range[0],range[1],0.05)
  scan_R = np.vectorize(lambda x: sim_groups(R_W=x,R_L=x,delta=0,rh=1,t_cycle=14,dt=0.1,n_groups=1,periods=[[[0,4]]],n_cont=k,L_days=L_days,I_days=I_days)[0,0])
  slopes = scan_R(R_range)
  s_to_R = interp1d(slopes,R_range,kind='slinear')
  R_to_s = interp1d(R_range,slopes,kind='slinear')
  return s_to_R#,slopes,R_range

s_r_conv_func = lambda r,s,g,m,n: (r*(r/(s*m)+1)**m)/(g*(1-(r/(g*n)+1)**(-n)))
R_func = lambda r,s,g,m,n: n*(1+r/(s*m))**m/(sum([(1+r/(g*n))**(-x) for x in range(1,n+1)]))
```

#### Figure S10 - the effect time-restriction and antiphasing

```python
s_r_conv = get_s_to_R(L_days=3,I_days=4,range=[0.1,3])
tot_effect = np.vectorize(lambda i,j: sim_groups(R_W=i,R_L=j,rh=1,t_cycle=14,dt=0.1,periods=[[[0,4]]],n_groups=1,delta=0,L_days=3,I_days=4,tmax=9*14)[0][0])
```

```python
tot_effect_15_06 = s_r_conv(tot_effect(1.5,0.6))
wa_R_15_06 = (4*1.5+0.6*10)/14
```

```python
res = 11
R_L_space = np.linspace(0,1,res)
R_W_space = np.linspace(1,3,res)
RLs, RWs = np.meshgrid(R_L_space,R_W_space)
wa_R = (4*RWs+RLs*10)/14
effect_res = tot_effect(RWs,RLs)
effect_R = s_r_conv(effect_res)
tot_reduction = RWs - effect_R
phase_reduction = wa_R-effect_R
phase_fraction = phase_reduction/tot_reduction
phase_fraction[0,-1] = 0
```

```python
# fig = init_fig(aspect=(8,5),fontsize=10)
fig,_= plt.subplots(1,2,figsize=(10,6),dpi=300)

ax1 = plt.subplot(1,2,1,aspect=2.5)
data1 = [wa_R_15_06-1.5,tot_effect_15_06-wa_R_15_06,tot_effect_15_06-1.5]
bottom = [1.5,wa_R_15_06,1.5]
plt.bar(range(3),data1,bottom=bottom,alpha=0.1,color=['r','g','k'])
plt.ylim([0.6,1.55])
plt.xlim([-0.5,3])
plt.yticks(np.arange(0.6,1.6,0.1))
plt.ylabel('reproduction number (R)')
plt.xticks(range(3),labels=['time-restriction','antiphasing','total effect'])
for i, v in enumerate(data1):
    ax1.text(i-0.2,v+bottom[i]-0.05, '{:.2f}'.format(v),fontsize=12)

ax1.text(2.45,tot_effect_15_06-0.02, '$R(t)$='+'{:.2f}'.format(tot_effect_15_06),fontsize=12,bbox=dict(facecolor='white', alpha=0.5))
plt.hlines(1.5,-0.4,2.4,linestyle='dashed')
ax1.text(0.45,1.48, '$R_W$=1.5',fontsize=12,bbox=dict(facecolor='white', alpha=1))
plt.hlines(1.5+data1[0],-0.4,1.4,linestyle='dashed')
plt.hlines(1.5+data1[2],0.6,2.4,linestyle='dashed')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax1.spines['left'].set_visible(False)
plt.arrow(0, 1.5-0.01, 0, data1[0]+0.03, head_width=0.04, head_length=0.02, linewidth=1, color='k', length_includes_head=True)    
plt.arrow(1, wa_R_15_06, 0, data1[1], head_width=0.04, head_length=0.02, linewidth=1, color='k', length_includes_head=True)    
plt.arrow(2, 1.5-0.01, 0, data1[2]+0.03, head_width=0.04, head_length=0.02, linewidth=1, color='k', length_includes_head=True)    
plt.title('relative contribution of antiphasing\nand time-restriction to the decrease of the\neffective reproduction number')
ax1.text(-0.1, 1.15, 'A', transform=ax1.transAxes,fontsize=16, va='top', ha='right')
ax1.text(0.55,0.9, 'antiphasing\naccounts for '+'{:.0f}%'.format(data1[1]/data1[2]*100)+'\nof total effect',fontsize=9,bbox=dict(facecolor='white', alpha=0.5))

ax2 = plt.subplot(1,2,2)
data = np.flipud(phase_fraction).T
plt.imshow(data,cmap='coolwarm')
# plt.clim([0,1])

plt.yticks(range(res),labels=['{:.1f}'.format(x) for x in RLs[0,:]]);
plt.xticks(range(res),labels=['{:.1f}'.format(x) for x in np.flipud(RWs)[:,0]]);
plt.ylabel('$R_L$')
plt.xlabel('$R_W$')
cbar = plt.colorbar(shrink=0.5)


cbar.set_label('fraction of R reduction caused by antiphasing')
plt.title('fraction of antiphasing effect out of\nthe total effect of the cyclic strategy as\na function of $R_W$ and $R_L$ ($T_{lat}$=3; $T_{inf}$=4; k=2)');

CS = plt.contour(data,colors='k')
plt.clabel(CS, inline=1, fontsize=10)
ax2.text(-0.1, 1.15, 'B', transform=ax2.transAxes,fontsize=16, va='top', ha='right')
plt.tight_layout()
plt.savefig('figS10_20200527.png')
```
